### Part 1: MC Prediction

In this section, you will write your own implementation of MC prediction (for estimating the action-value function).  

Investigate a policy where the player _almost_ always sticks if the sum of her cards exceeds 18.  In particular, she selects action `STICK` with 80% probability if the sum is greater than 18; and, if the sum is 18 or below, she selects action `HIT` with 80% probability.  The function `generate_episode_from_limit_stochastic` samples an episode using this policy. 

The function accepts as **input**:
- `bj_env`: This is an instance of OpenAI Gym's Blackjack environment.

It returns as **output**:
- `episode`: This is a list of (state, action, reward) tuples (of tuples) and corresponds to $(S_0, A_0, R_1, \ldots, S_{T-1}, A_{T-1}, R_{T})$, where $T$ is the final time step.  In particular, `episode[i]` returns $(S_i, A_i, R_{i+1})$, and `episode[i][0]`, `episode[i][1]`, and `episode[i][2]` return $S_i$, $A_i$, and $R_{i+1}$, respectively.

````python
def generate_episode_from_limit_stochastic(bj_env):
    episode = []
    state = bj_env.reset()
    while True:
        probs = [0.8, 0.2] if state[0] > 18 else [0.2, 0.8]
        action = np.random.choice(np.arange(2), p=probs)
        next_state, reward, done, info = bj_env.step(action)
        episode.append((state, action, reward))
        state = next_state
        if done:
            break
    return episode
````

Write your own implementation of MC prediction.  Feel free to implement either first-visit or every-visit MC prediction; in the case of the Blackjack environment, the techniques are equivalent.

Algorithm has three arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `generate_episode`: This is a function that returns an episode of interaction.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

````python
def mc_prediction_q(env, num_episodes, generate_episode, gamma=1.0):
    # initialize empty dictionaries of arrays
    returns_sum = defaultdict(lambda: np.zeros(env.action_space.n))
    N = defaultdict(lambda: np.zeros(env.action_space.n))
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
         
        # generate an episode
        episode = generate_episode(env)
        # obtain the state, action and rewards
        states, actions, rewards = zip(*episode)
        #discount
        discounts = np.array([gamma**i for i in range(len(rewards)+1)])
        #estimation for each state-action pair in episode
        for i, state in enumerate(states):
            #update  the return sum 
            returns_sum[state][actions[i]] += sum(rewards[i:]* discounts[:-(1+i)])
            #update the num of episodes 
            N[state][actions[i]] += 1.0
            Q[state][actions[i]] = returns_sum[state][actions[i]]/N[state][actions[i]]
    return Q
````

Obtain the action-value function estimate $Q$.  Plotted the corresponding state-value function.

### Part 2: MC Control

Write your own implementation of constant-$\alpha$ MC control.  

The algorithm must have four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.
- `policy`: This is a dictionary where `policy[s]` returns the action that the agent chooses after observing state `s`.


````python
def generate_episode_from_Q(env, Q, epsilon, nA): #generate episode following the epsilon greedy 
    episode = []
    state = env.reset()
    while True:
        action = np.random.choice(np.arange(nA), p=probability(Q[state], epsilon, nA)) \
                                                    if state in Q else env.action_space.sample()
       
        next_state, reward, done, info = env.step(action)
        episode.append((state, action, reward))
        state = next_state 
        if done:
            break
    return episode


def probability(Q_s, epsilon , nA ): #constructing the policy pi that is e_greedy with respect to the currect action-value function estimate Q
    
    policy_pi = np.ones(nA) * epsilon / nA
    best_action = np.argmax(Q_s)
    policy_pi[best_action] = 1- epsilon + (epsilon/nA)
    return policy_pi
    
def update_Q (env, episode, Q, gamma, alpha): #update the estimate action-value after each episode
    
    states, actions, rewards = zip(*episode)
    
    discounts = np.array([gamma**i for i in range(len(rewards)+1)])
    for i, state in enumerate(states):
    
        old_Q = Q[state][actions[i]]
        Q[state][actions[i]] = old_Q + alpha * (sum(rewards[i:]*discounts[:-(i+1)])- old_Q)
    return Q
````

````python
def mc_control(env, num_episodes, alpha, gamma=1.0, eps_start =1.0, eps_decay = 0.9999, eps_min = 0.03):
    nA = env.action_space.n
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(nA))
    epsilon = eps_start
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        #set the value of epsilon
        epsilon = max(epsilon*eps_decay, eps_min)
        #generate an episode (policy evaluation)
        episode = generate_episode_from_Q(env, Q, epsilon, nA)
        
        #update the action-value function 
        Q = update_Q(env, episode, Q, gamma, alpha)
    #take the policy (policy improvement), taking the action that maximizes the corresponded row in that action
    policy = dict((k, np.argmax(v)) for k,v in Q.items())
        
    return policy, Q
````
